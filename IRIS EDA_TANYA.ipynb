```python
url = 'https://raw.githubusercontent.com/theanalytix/DAInterns-CIEI/main/Datasets/Iris_data.csv'
df = pd.read_csv(url)
```

```python
#counting the species
df['Species'].value_counts()
```

```python
sns.set_theme()
plt.figure(figsize=(12,7))
sns.countplot(x=df['Species'],palette='viridis')
plt.title('The Species')
```

```python
#Analysing species wrt to individual numerical categories of the data frame
plt.figure(figsize=(18,10))
sns.set_theme()
plt.subplot(1,2,1)
sns.swarmplot(x='Species',y='SepalLengthCm',data=df)
plt.title('Sepal Length')
plt.subplot(1,2,2)
sns.swarmplot(x='Species',y='SepalWidthCm',data=df)
plt.title('Sepal Width')
```

```python
plt.figure(figsize=(18,10))
sns.set_theme()
plt.subplot(1,2,1)
sns.stripplot(x='Species',y='PetalLengthCm',data=df,jitter=True)
plt.title('Petal Length')
plt.subplot(1,2,2)
sns.stripplot(x='Species',y='PetalWidthCm',data=df,jitter=True)
plt.title('Petal Width')
```

```python
#Analysing the numerical categories of the dataframe
plt.figure(figsize=(18,10))
sns.set_theme()
plt.subplot(1,2,1)
sns.kdeplot(x='SepalLengthCm',y='SepalWidthCm',data=df,shade=True)
plt.title('Overall concentration of values-Sepal length and Width')
plt.subplot(1,2,2)
sns.kdeplot(x='PetalLengthCm',y='PetalWidthCm',data=df,shade=True)
plt.title('Overall concentration of values-Petal length and width')
```

```python
#Relationship between sepal length and width, and petal length and width wrt species
plt.figure(figsize=(18,10))
sns.set_theme()
plt.subplot(1,2,1)
sns.scatterplot(x='SepalLengthCm',y='SepalWidthCm',data=df,hue='Species')
plt.title('Relationship b/w sepal length and width wrt species')
plt.subplot(1,2,2)
sns.scatterplot(x='PetalLengthCm',y='PetalWidthCm',data=df,hue='Species')
plt.title('Relationship b/w petal length and width wrt species')
```

```python
#Relationship between sepal length and width, and petal length and width 
plt.figure(figsize=(18,10))
sns.set_theme()
plt.subplot(1,2,1)
sns.scatterplot(x='SepalLengthCm',y='SepalWidthCm',data=df)
plt.title('Relationship b/w sepal length and width')
plt.subplot(1,2,2)
sns.scatterplot(x='PetalLengthCm',y='PetalWidthCm',data=df)
plt.title('Relationship b/w petal length and width')
```

```python
#Finding the correlation b/w the numeric columns
df1 = df.drop(['Id'],axis=1)
cor = df1.corr()
sns.heatmap(cor,annot=True)
```

```python
#Establishing a comparison between ALL the numerical categories of the dataframe
sns.set_theme()
plt.figure(figsize=(14,10))
sns.pairplot(df,hue='Species')
```
